```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import numpy as np
import time
import os.path

pd.set_option('display.max_rows', 500)
```

```python
def load_bmc_results(db_name, table_name):
    print("File {} last modified: {}".format(db_name, time.ctime(os.path.getmtime(db_name))))
    conn = sqlite3.connect(db_name)
    df = pd.read_sql_query("select * from " + table_name + ";", conn)
    return df
```

```python
# configuration

bound = 4
only_sat = False
only_unsat = False
file_prefix = ""
```

```python
df_A = load_bmc_results("bmc-phi-" + str(bound) + ".sqlite", "bmc_" + str(bound))
df_A
```

```python
df_B = load_bmc_results("bmc-gsa-" + str(bound) + ".sqlite", "bmc_" + str(bound))
df_B
```

```python
def sanity_check(A, B):
    print(len(A), 'rows')
    assert len(A) == len(B)
    zipped = zip(list(A.itertuples(index=True)), list(B.itertuples(index=True)))
    
    seen_files = set()
    
    for row in zipped:
        row_A = row[0]
        row_B = row[1]
        file_A = getattr(row_A, 'file')
        file_B = getattr(row_B, 'file')
        assert file_A == file_B
        
        assert file_A not in seen_files
        seen_files.add(file_A)
        
        res_A = getattr(row_A, 'result')
        res_B = getattr(row_B, 'result')
        if res_A != '' and res_B != '':
            assert res_A == res_B
        
        res_bmc_A = getattr(row_A, 'BMC_result')
        res_bmc_B = getattr(row_B, 'BMC_result')
        if res_bmc_A != '' and res_bmc_B != '':
            assert res_bmc_A == res_bmc_B
        
    print('Sanity checks succeeded')
        
sanity_check(df_A, df_B)
```

```python
def skip_crashes(df):
    return df.query('status == 0').query('result != ""')

def skip_sat(df):
    return df.query('result != "sat"')

def skip_unsat(df):
    return df.query('result != "unsat"')
```

In Jupyter it is much prettier to return an object with the values and implement `_repr_html` method on it to display it conveniently. Printing a table is not the way to go.

```python
def print_summary(df, name):
    num_runs = len(df)
    num_finished = len(skip_crashes(df))
    num_sat = len(df.query('result == "sat"'))
    num_unsat = len(df.query('result == "unsat"'))
    
    print("\n~~~ Results for " + name + ": ~~~")
    print("\tno. runs:", num_runs, "\tno. finished:", num_finished,
          "\n\tno. sat:", num_sat, "\tno. unsat:", num_unsat)
    

print_summary(df_A, "A")
print_summary(df_B, "B")
```

```python
def preprocess(df, name):
    res = skip_crashes(df)
    
    assert not (only_sat and only_unsat), "Contradictory config. Choose one."
    
    if only_sat:
        res = res.query('result == "sat"')
    if only_unsat:
        res = res.query('result == "unsat"')
    if file_prefix is not None:
        res = res[res.file.str.startswith(file_prefix)]
        
    print_summary(res, name)
    return res
```

```python
succeeded_A = preprocess(df_A, "A")
succeeded_A
```

```python
succeeded_B = preprocess(df_B, "B")
succeeded_B
```

```python
list(zip(list(succeeded_B.index), list(succeeded_B.file)))
```

```python
def compare_values(columns, df_A, df_B):
    vals_A = df_A[['file'] + list(columns)]
    vals_B = df_B[['file'] + list(columns)]
    merged = vals_A.merge(vals_B, how='inner', left_index=True, on='file')
    return merged

def scatter_values(column, df_A, df_B):
    vals = compare_values((column, ), df_A, df_B)
#     res = vals.plot.scatter(x = column + "_x", y = column + "_y")
    xlabel = column + '_x'
    ylabel = column + '_y'
    
    xs = vals[xlabel]
    ys = vals[ylabel]
    
    print('len(A):', len(df_A), '\tlen(B):', len(df_B))
    print('No. values', len(xs))
    print('Min val, A:', min(xs), '\tB:', min(ys))
    print('Max val, A:', max(xs), '\tB:', max(ys))
    print('Avg val, A:', "{:0.2f}".format(np.mean(xs)), '\tB:', "{:0.2f}".format(np.mean(ys)))
    
    min_val = min(0, min(xs), min(ys))
    max_val = max(max(xs), max(ys))
    plt.xlim(min_val, max_val)
    plt.ylim(min_val, max_val)
    
    plt.scatter(xs, ys)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xlabel(column + '_A')
    plt.ylabel(column + '_B')
    
    mid_x = np.linspace(min_val, max_val, 2)
    mid_y = mid_x
    plt.plot(mid_x, mid_y, 'r-')
    plt.title(column)
    
    plt.rcParams['figure.dpi'] = 160
    plt.draw()
    
```
